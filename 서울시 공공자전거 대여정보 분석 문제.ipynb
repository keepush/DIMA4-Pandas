## 대여소 정보 데이터 읽기

```python
import numpy as np
import pandas as pd

df = pd.read_excel('서울특별시 공공자전거 대여소 정보(19.12.9).xlsx')
```

## NA 값이 몇 개인지 확인

- NA값이 들어있는 데이터 삭제

```python
display(df.isna().sum())
```

```python
# 결측치 확인
display(df[df['대여소ID'].isna()])
```

```python
df = df.dropna()
# df.dropna(inplcae=True)  -  결측치 가지는 모든 행 삭제
```

## 타입 변환

- 대여소ID ==> int32
- 기준시작일자 ==> 날짜타입

```python
df['대여소ID'] = df['대여소ID'].astype('int32')
```

```python
print(df.info())
```

```python
df2 = df.copy()
```

```python
df2['기준시작일자'] = pd.to_datetime(df2['기준시작일자'])
```

```python
print(df2['기준시작일자'].str.contains('개통').sum())   # contains = T/F로 결과 반환
```

```python
df2['기준시작일자'] = df2['기준시작일자'].replace('개통','', regex = True)
print(df2['기준시작일자'].str.contains('개통').sum())
```

```python
print(df2.info())
```

```python
display(df2.head(10))
```

```python
df = df2
```

## 대여소_구의 목록, 개수 출력

```python
print(len(df['대여소_구'].unique()))
```

```python
print(df['대여소_구'].value_counts())
```

## 대여소_구별 거치대 수

```python
place = df.groupby('대여소_구')['거치대수'].sum().sort_values(ascending=False)
print(place)        # Series 객체이기 떄문에 sort_values에 기준 지정 X
```

## 구별 거치대수의 합계, 평균, 최대, 최소 출력하고 거치대가 많은 순으로 합계별 내림차순으로 10개 출력

```python
result = df.groupby('대여소_구')['거치대수'].agg(['sum','mean','max','min']).sort_values('sum', ascending=False).head(10).round(2)
display(result)
```

```python
result = df.groupby('대여소_구')['거치대수'].agg(['sum','mean','max','min'])
print(result.sort_values(result.columns[0],ascending=False).head(10))
```

# 서울시 공공자전거 대여 정보 분석
- 대여정보 읽기
- 대여정보 확인 (결측치 컬럼 등)
- 대여일시, 반납일시를 날짜타입으로 변경
- 일자별로 대여 건수 내림차순으로 정렬하여 출력
    - Series객체.dt.날짜property
- 요일별 대여 건수 확인

## 대여 정보 데이터 읽기

```python
rent = pd.read_csv('서울특별시 공공자전거 대여정보_201911_2.csv', encoding='cp949')
```

## 대여정보 확인

```python
print(rent.info())
print(rent.isna().sum())
```

## 대여일시, 반납일시를 날짜타입으로 변경

```python
rent['대여일시'] = pd.to_datetime(rent['대여일시'])
rent['반납일시'] = pd.to_datetime(rent['반납일시'])
print(rent.info())
```

```python
display(rent.head(10))
```

## 요일별 대여 건수 확인

```python
rent['요일'] = rent['대여일시'].dt.dayofweek          # Series 객체일 경우 dt 사용해야 datetime 객체 가져오기 可
print(rent.head(5))
```

```python
# Series 값을 특정 값으로 바꿀 경우, .map() 사용
dayofweek = {0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'}
rent['요일'] = rent['요일'].map(dayofweek)
display(rent.head(3))
```

```python
result = rent.groupby('요일')['요일'].count().sort_values(ascending=False).apply(lambda x:f'{x:,}건')
print(result)
```

## 일자별로 대여 건수 내림차순으로 정렬하여 출력

```python
# 일자에 대한 정보 새로운 컬럼으로 생성
rent['day'] = rent['대여일시'].dt.day
result = rent.groupby('day')['day'].count().sort_values(ascending=False)
print(result)
```

# 대여소 정보와 대여정보 Join
- 대여소 정보와 대여정보 join
- join 데이터 정보 확인
- 각 구별 이용시간과  이용거리에 대한 평균, 최대, 최소값 출력

```python
print(rent.info())
```

## 대여소 ID 기준으로 두 데이터 병합

```python
rent_bicycle = pd.merge(df, rent, left_on='대여소ID', right_on='대여 대여소번호')
```

```python
rent_bicycle.drop('대여 대여소번호', inplace=True, axis=1)
```

## 구별 평균 이용시간, 평균 이용거리, 최대 이용시간, 최소 이용시간 확인 (구 이름 정렬)

```python
result = rent_bicycle.groupby('대여소_구')[['이용시간','이용거리']].agg(['mean','max','min']).sort_values(('이용시간','mean'), ascending=False)
display(result)
```
